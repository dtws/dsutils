# インポート

```python
from dtws1 import heatmap
from h3 import h3
import numpy as np
import pandas as pd
```

# データ準備

虎ノ門ヒルズの周辺のH3メッシュ

```python
hills_lat, hills_lng = 35.66687568, 139.74947495
ids = list(h3.k_ring(h3.geo_to_h3(hills_lat, hills_lng, 10), 10))
print(ids[:3])
```

H3メッシュの中心点の緯度・経度

```python
lats = [h3.h3_to_geo(i)[0] for i in ids]
lngs = [h3.h3_to_geo(i)[1] for i in ids]
print(lats[0], lngs[0])
```

データフレームにまとめる 

```python
df = pd.DataFrame(dict(
    id = ids,
    lat = lats,
    lng = lngs
))
df["rand"] = np.random.randint(1,15,size=len(df)) ## Random numbers are here to exhibit different colors of grids
df.head()
```

# 例

```python
fmap1 = heatmap.draw_folium(
    df=df, 
    id_col="id",
    val_col="rand", 
    fill_opacity=0.8, # 透明度（色塗り）
    line_opacity=0.2, # 透明度（境界）
    zoom_start=15
)
fmap1
```
